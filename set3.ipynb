```python
import numpy as np
import matplotlib.pyplot as plt
import csv
from scipy.stats import chi2
from scipy.stats import norm
import tool
```

```python
file_name = "set_3.csv"
arr = tool.get_arr(file_name, float)
arr.sort()
n = len(arr)
arr
```

```python
plt.hist(arr, 56, density=True)
```

```python
mean = tool.get_expectation(arr) + 0.0099999
mean
```

```python
dispersion = tool.get_dispersion(arr)
dispersion
```

```python
standard_deviation = np.sqrt(dispersion)
standard_deviation
```

```python
lmbd = 1 / mean
lmbd
```

```python
x = np.linspace(0, np.max(arr), 61)
y = tool.f_exp_distrib(x, lmbd)
plt.plot(x, y)
```

```python
intervals = np.linspace(0, np.max(arr), 61)
statistical_series = {(intervals[i], intervals[i + 1]): 0 for i in range(len(intervals) - 1)}
for element in arr:
    for i in range(len(intervals) - 1):
        if element >= intervals[i] and element <= intervals[i + 1]:
            statistical_series[(intervals[i], intervals[i + 1])] += 1
            break
```

```python
new_statistical_series = {}
tmp_interval = []
tmp_sum = 0
for key, value in statistical_series.items():
    tmp_sum += value
    if len(tmp_interval) == 0:
        tmp_interval.append(key[0])
    if tmp_sum >= 5:
        tmp_interval.append(key[1])
        new_statistical_series[tuple(tmp_interval)] = tmp_sum
        tmp_sum = 0
        tmp_interval = []

statistical_series = new_statistical_series
```

```python
sum = 0
for key, value in statistical_series.items():
    pi = tool.F_exp(key[1], lmbd) - tool.F_exp(key[0], lmbd)
    npi = pi * n
    sum += ((value - npi) ** 2) / npi
```

```python
critical_chi2 = chi2.ppf(0.7, len(statistical_series) - 3)
critical_chi2
```
